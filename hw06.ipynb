```python
import kinematics as kin
import transforms as tr
from visualization import VizScene
import numpy as np
import time
np.set_printoptions(precision=4, suppress=True)
```

```python
#dh order: theta, d, a, alpha
dh = np.array([[0, 0, 1, np.pi/2],
               [0, 0, 1, 0]])

jt_types = ['r', 'r']

arm = kin.SerialArm(dh, jt=jt_types)

q_set_0 = [0, 0]
q_set_1 = [0, np.pi/4]
q_set_2 = [0, np.pi/2]
q_set_3 = [np.pi/4, np.pi/4]
q_set_4 = [0, 0]
q_set_5 = [0, 0]

J0 = arm.jacob(q_set_0)
J1 = arm.jacob(q_set_1)
J2 = arm.jacob(q_set_2)
J3 = arm.jacob(q_set_3)
J4 = arm.jacob(q_set_4)
J5 = arm.jacob(q_set_5)

viz = VizScene()
viz.add_arm(arm, draw_frames=True)
viz.update(qs=[q_set_0])
viz.hold()

viz = VizScene()
viz.add_arm(arm, draw_frames=True)
viz.update(qs=[q_set_1])
viz.hold()

viz = VizScene()
viz.add_arm(arm, draw_frames=True)
viz.update(qs=[q_set_2])
viz.hold()

viz = VizScene()
viz.add_arm(arm, draw_frames=True)
viz.update(qs=[q_set_3])
viz.hold()

viz = VizScene()
viz.add_arm(arm, draw_frames=True)
viz.update(qs=[q_set_4])
viz.hold()

viz = VizScene()
viz.add_arm(arm, draw_frames=True)
viz.update(qs=[q_set_5])
viz.hold()
```

```python
tau_1 = J1.T @ np.array([-1,0,0,0,0,0]).T
print("Joint Torque 1")
print(tau_1)
tau_2 = J2.T @ np.array([-1,0,0,0,0,0]).T
print("Joint Torque 2")
print(tau_2)
tau_3 = J3.T @ np.array([-1,-1,0,0,0,0]).T
print("Joint Torque 3")
print(tau_3)
tau_4 = J4.T @ np.array([0,0,1,0,0,0]).T
print("Joint Torque 4")
print(tau_4)
tau_5 = J5.T @ np.array([1,0,0,0,0,0]).T
print("Joint Torque 5")
print(tau_5)
```
